```python

import pandas as pd
import matplotlib as mb
import matplotlib.pyplot as plt
xls = pd.read_csv('_.-tag-loss.csv')
xls1 = pd.read_csv('_.-tag-val_loss.csv')

# 原点相交
ax = plt.gca()
# ax.xaxis.set_ticks_position('bottom')
# ax.spines['bottom'].set_position(('data',0))
# ax.yaxis.set_ticks_position('left')
# ax.spines['left'].set_position(('data',0))
# 去掉边框
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
plt.plot(xls.Step, xls.Value, label='loss')
# plt.scatter(xls.Step, xls.Value)
plt.plot(xls1.Step, xls1.Value, label='val_loss')
plt.title('ResNeXt with focal loss',fontsize='large') 
plt.legend()

plt.savefig('ResNeXt_loss.png', dpi=300)
plt.show()
```
